```python
headlines=[]
for row in range(0, len(data)):
  headlines.append(' '.join([str(i) for i in data.iloc[row, 0:25]]))
```

```python
headlines[0:3]
```

```python
# implement bag of words
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
cv = CountVectorizer(ngram_range=(2, 2))
traindata_x = cv.fit_transform(headlines)
```

```python
traindata_x
```

```python
train.Label.unique()
```

```python
randomclassifier = RandomForestClassifier(n_estimators=200, criterion='entropy')
randomclassifier.fit(traindata_x, train.Label)
```

```python
test_transform=[]
for row in range(0, len(test)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row, 2:]))
```

```python
test_data = cv.transform(test_transform)
```

```python
randomclassifier.fit(traindata_x, train.Label)
```

```python
pred = randomclassifier.predict(test_data)
```

```python
# plot confusion matrix
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
import seaborn as sns

def plot_cm(cm, title='Confusion Matrix', labels=['Positive', 'Negative']):
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, cmap='Blues', fmt='d', xticklabels=labels, yticklabels=labels)
    plt.title(title)
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()
```

```python
plt.figure()
cm = confusion_matrix(test.Label, pred)
plot_cm(cm)
```

```python
score=accuracy_score(test['Label'], pred)
print(score)
report=classification_report(test['Label'], pred)
print(report)
```

```python
# use multinomialNB
```

```python
from sklearn.naive_bayes import MultinomialNB
nb=MultinomialNB()
nb.fit(traindata_x,train['Label'])

predictions = nb.predict(test_data)
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
score=accuracy_score(test['Label'],predictions)
print(score)
report=classification_report(test['Label'],predictions)
print(report)
```
